# Règles du Jeu d'Échecs

## Objectif du Jeu
Le but du jeu est de mettre l'adversaire en échec et mat, c'est-à-dire de placer son roi sous une attaque directe sans qu'il puisse être défendu, évitant ainsi une capture inévitable au prochain tour.

## L'Échiquier
- **Dimensions** : L'échiquier est un plateau carré de 8x8 cases alternant entre clair et sombre.
- **Disposition** : Les pièces sont disposées de la manière suivante :
  - **Rangée 1 et 8 (depuis la perspective de chaque joueur)** : Tour, Cavalier, Fou, Reine, Roi, Fou, Cavalier, Tour.
  - **Rangée 2 et 7** : 8 pions.

## Les Pièces et leurs Mouvements
### Le Roi
- **Mouvement** : Il se déplace d'une case dans n'importe quelle direction (horizontalement, verticalement, ou en diagonale).
- **Règles spéciales** :
  - **Roque** : Le roi peut effectuer un mouvement spécial appelé le "roque" en coopération avec l'une des tours. Cela implique de déplacer le roi de deux cases vers la tour, et la tour de se positionner immédiatement à côté du roi de l'autre côté. Les conditions pour effectuer un roque sont les suivantes :
    1. Ni le roi ni la tour n'ont bougé précédemment.
    2. Aucune pièce ne se trouve entre le roi et la tour.
    3. Le roi ne doit pas être en échec, ne doit pas traverser une case attaquée, ni finir sur une case attaquée.
    4. Le roque peut être petit (côté roi) ou grand (côté dame).

### La Reine
- **Mouvement** : La reine peut se déplacer de n'importe quel nombre de cases en ligne droite, que ce soit horizontalement, verticalement ou en diagonale.
- **Règles spéciales** : La reine n'a pas de mouvement spécial au-delà de son grand rayon d'action, mais sa polyvalence la rend essentielle pour contrôler l'échiquier et soutenir les autres pièces.

### La Tour
- **Mouvement** : La tour se déplace de n'importe quel nombre de cases en ligne droite, horizontalement ou verticalement.
- **Règles spéciales** :
  - **Roque** : Comme mentionné plus haut, la tour participe au mouvement spécial du roque avec le roi. Lors du roque, la tour se déplace pour se positionner juste à côté du roi, après que celui-ci ait bougé de deux cases.

### Le Fou
- **Mouvement** : Le fou se déplace de n'importe quel nombre de cases en diagonale.
- **Règles spéciales** : Le fou est limité à une seule couleur (soit toujours sur les cases claires, soit sur les cases sombres), ce qui le rend complémentaire à l'autre fou dans la stratégie de contrôle des diagonales.

### Le Cavalier
- **Mouvement** : Le cavalier se déplace en formant un "L" : deux cases dans une direction (horizontalement ou verticalement) puis une case perpendiculairement, ou vice versa.
- **Règles spéciales** :
  - **Sauter par-dessus les pièces** : Le cavalier est la seule pièce capable de sauter par-dessus les autres pièces, qu'elles soient alliées ou ennemies. Cela lui permet d'atteindre des cases normalement inaccessibles par d'autres pièces.

### Le Pion
- **Mouvement** : Le pion avance d'une case en ligne droite. Lors de son premier mouvement, il peut avancer de deux cases. Les pions capturent en diagonale.
- **Règles spéciales** :
  - **Promotion** : Lorsqu'un pion atteint la dernière rangée de l'échiquier, il est promu en une autre pièce de son choix : une reine, une tour, un fou ou un cavalier, indépendamment des pièces déjà présentes sur l'échiquier. La promotion en reine est le choix le plus fréquent en raison de sa puissance, mais dans certaines situations stratégiques, une autre promotion peut être avantageuse.
  - **Prise en passant** : Si un pion avance de deux cases depuis sa position initiale et se retrouve à côté d'un pion adverse, ce dernier peut capturer ce pion "en passant" en se déplaçant sur la case que le pion aurait occupée s'il n'avait avancé que d'une case. Cette capture doit être effectuée immédiatement après le double pas du pion adverse.

## Concepts Clés
### Échec
Lorsqu'un roi est attaqué, il est en échec. Le joueur doit alors protéger son roi en se déplaçant, en capturant la pièce attaquante ou en interposant une pièce.

### Échec et Mat
L'échec et mat survient lorsque le roi est en échec et ne peut pas se soustraire à l'attaque. Cela signifie la fin de la partie et la victoire de l'adversaire.

### Roque
- **Petit Roque** : Le roi se déplace de deux cases vers la droite, et la tour à sa droite vient se placer juste à côté du roi.
- **Grand Roque** : Le roi se déplace de deux cases vers la gauche, et la tour à sa gauche vient se placer juste à côté du roi.
- **Conditions** : Le roi et la tour concernés ne doivent pas avoir encore bougé, et aucune pièce ne doit se trouver entre eux. Le roi ne peut pas être en échec, ni traverser ou atterrir sur une case attaquée.

### Pat
Un pat se produit lorsqu'un joueur n'a plus de coup légal et que son roi n'est pas en échec. La partie se termine par un match nul.

### Match Nul
Il y a plusieurs façons d'arriver à un match nul :
- Par accord mutuel entre les joueurs.
- Par position répétée trois fois.
- Par manque de matériel suffisant pour mettre l'adversaire en échec et mat.
- Par règle des 50 coups, où aucun pion n'a été déplacé et aucune pièce n'a été capturée durant les 50 derniers coups.

## Fonctionnalités du Jeu

### Lancement d'une Partie
- Le jeu commence immédiatement après l'exécution du script. L'échiquier s'affiche, et c'est au tour des **bleus** de jouer en premier.

### Entrée des Mouvements
- **Format des Mouvements** : Entrez vos coups dans le format 'e2 e4' où la première partie ('e2') est la case de départ et la seconde ('e4') est la case d'arrivée.

### Promotion des Pions
- Lorsque votre pion atteint la dernière rangée, le script vous demande en quoi vous souhaitez promouvoir votre pion : **Q** pour Reine, **R** pour Tour, **B** pour Fou, **C** pour Cavalier.

### Sauvegarde de la Partie
- Vous pouvez sauvegarder la partie en cours en appelant la méthode `save_game(filename)`. Le nom du fichier est donné sous la forme d'une chaîne de caractères.

### Chargement d'une Partie
- Pour charger une partie précédemment sauvegardée, appelez la méthode `load_game(filename)` en fournissant le nom du fichier.

### Conditions de Fin de Partie
- **Échec et Mat** : Si un joueur ne peut éviter l'échec, la partie se termine et l'autre joueur est déclaré vainqueur.
- **Pat** : Si un joueur ne peut pas effectuer de mouvement légal et n'est pas en échec, la partie se termine en match nul.
- **Match Nul** : Le jeu se termine également par un match nul selon les règles établies (ex. par accord mutuel ou la règle des 50 coups).

### Historique et Résumé de la Partie
- À la fin de la partie, un résumé des mouvements ainsi que le temps écoulé pour chaque joueur est affiché.

````python
from IPython.display import clear_output, display
import os
import time

class Piece:
    def __init__(self, color):
        self.color = color
        self.has_moved = False
    
    def can_move(self, board, start, end):
        raise NotImplementedError("Cette méthode doit être implémentée par les sous-classes")

class Pawn(Piece):
    def can_move(self, board, start, end):
        direction = 1 if self.color == 'red' else -1
        start_row, start_col = start
        end_row, end_col = end

        if start_col == end_col:  # Avancer
            if (end_row - start_row) == direction:
                return board[end_row][end_col] is None
            if (start_row == 1 and self.color == 'red') or (start_row == 6 and self.color == 'blue'):
                return (end_row - start_row) == 2 * direction and board[end_row][end_col] is None and board[start_row + direction][start_col] is None
        elif abs(start_col - end_col) == 1 and (end_row - start_row) == direction:  # Capturer
            if board[end_row][end_col] is not None:
                return board[end_row][end_col].color != self.color
            # Prise en passant
            if isinstance(board[start_row][end_col], Pawn) and board[start_row][end_col].color != self.color:
                if (start_row == 4 and self.color == 'red') or (start_row == 3 and self.color == 'blue'):
                    return True
        
        return False

    def promote(self, board, end):
        row, col = end
        if (self.color == 'red' and row == 7) or (self.color == 'blue' and row == 0):
            while True:
                promotion = input("Promouvoir à (Q, R, B, C) : ").upper()
                if promotion in ['Q', 'R', 'B', 'C']:
                    if promotion == 'Q':
                        board[row][col] = Queen(self.color)
                    elif promotion == 'R':
                        board[row][col] = Rook(self.color)
                    elif promotion == 'B':
                        board[row][col] = Bishop(self.color)
                    elif promotion == 'C':
                        board[row][col] = Cavalier(self.color)
                    break

class Rook(Piece):
    def can_move(self, board, start, end):
        start_row, start_col = start
        end_row, end_col = end

        if start_row == end_row:  # Mouvement horizontal
            step = 1 if start_col < end_col else -1
            for col in range(start_col + step, end_col, step):
                if board[start_row][col] is not None:
                    return False
        elif start_col == end_col:  # Mouvement vertical
            step = 1 if start_row < end_row else -1
            for row in range(start_row + step, end_row, step):
                if board[row][start_col] is not None:
                    return False
        else:
            return False
        
        return board[end_row][end_col] is None or board[end_row][end_col].color != self.color

class Cavalier(Piece):
    def can_move(self, board, start, end):
        start_row, start_col = start
        end_row, end_col = end
        row_diff = abs(end_row - start_row)
        col_diff = abs(end_col - start_col)
        return (row_diff, col_diff) in [(2, 1), (1, 2)] and (board[end_row][end_col] is None or board[end_row][end_col].color != self.color)

class Bishop(Piece):
    def can_move(self, board, start, end):
        start_row, start_col = start
        end_row, end_col = end

        if abs(start_row - end_row) == abs(start_col - end_col):  # Mouvement diagonal
            row_step = 1 if start_row < end_row else -1
            col_step = 1 if start_col < end_col else -1
            for i in range(1, abs(end_row - start_row)):
                if board[start_row + i * row_step][start_col + i * col_step] is not None:
                    return False
        else:
            return False
        
        return board[end_row][end_col] is None or board[end_row][end_col].color != self.color

class Queen(Piece):
    def can_move(self, board, start, end):
        return Rook(self.color).can_move(board, start, end) or Bishop(self.color).can_move(board, start, end)

class King(Piece):
    def can_move(self, board, start, end):
        start_row, start_col = start
        end_row, end_col = end

        # Déplacement standard du roi
        if max(abs(start_row - end_row), abs(start_col - end_col)) == 1:
            return board[end_row][end_col] is None or board[end_row][end_col].color != self.color
        
        # Roque
        if not self.has_moved and start_row == end_row:
            if end_col == start_col + 2:  # Petit roque
                return isinstance(board[start_row][7], Rook) and not board[start_row][7].has_moved and \
                       board[start_row][5] is None and board[start_row][6] is None
            if end_col == start_col - 2:  # Grand roque
                return isinstance(board[start_row][0], Rook) and not board[start_row][0].has_moved and \
                       board[start_row][1] is None and board[start_row][2] is None and board[start_row][3] is None
        return False

    def castle(self, board, start, end):
        if end[1] == start[1] + 2:  # Petit roque
            board[start[0]][5], board[start[0]][7] = board[start[0]][7], None
        elif end[1] == start[1] - 2:  # Grand roque
            board[start[0]][3], board[start[0]][0] = board[start[0]][0], None

class Board:
    def __init__(self):
        self.board = self.create_board()
        self.en_passant_target = None
        self.move_history = []
        self.turn = 'blue'
        self.timer = Timer()
        self.player_times = {'red': Timer(), 'blue': Timer()}

    def create_board(self):
        board = [[None for _ in range(8)] for _ in range(8)]

        # Placer les pions
        for col in range(8):
            board[1][col] = Pawn('red')
            board[6][col] = Pawn('blue')

        # Placer les autres pièces
        for i, piece in enumerate([Rook, Cavalier, Bishop, Queen, King, Bishop, Cavalier, Rook]):
            board[0][i] = piece('red')
            board[7][i] = piece('blue')
        
        return board

    def move(self, start, end):
        piece = self.board[start[0]][start[1]]
        if piece and piece.can_move(self.board, start, end):
            if isinstance(piece, Pawn) and abs(start[1] - end[1]) == 1 and self.en_passant_target == end:
                self.board[start[0]][end[1]] = None  # Capturer en passant
            if isinstance(piece, King):
                piece.castle(self.board, start, end)
            self.board[end[0]][end[1]] = piece
            self.board[start[0]][start[1]] = None
            piece.has_moved = True

            if isinstance(piece, Pawn):
                if abs(start[0] - end[0]) == 2:  # Définir la cible de la prise en passant
                    self.en_passant_target = ((start[0] + end[0]) // 2, start[1])
                else:
                    self.en_passant_target = None
                piece.promote(self.board, end)
            else:
                self.en_passant_target = None

            self.move_history.append((start, end))
            self.turn = 'blue' if self.turn == 'red' else 'red'
        else:
            raise ValueError("Mouvement invalide")

    def is_in_check(self, color):
        king_position = None
        for row in range(8):
            for col in range(8):
                piece = self.board[row][col]
                if isinstance(piece, King) and piece.color == color:
                    king_position = (row, col)
                    break
            if king_position:
                break
        
        if not king_position:
            return False
    
        for row in range(8):
            for col in range(8):
                piece = self.board[row][col]
                if piece and piece.color != color:
                    if piece.can_move(self.board, (row, col), king_position):
                        return True
        return False

    def is_checkmate(self):
        return self.is_in_check(self.turn) and not self.get_all_legal_moves(self.turn)

    def is_stalemate(self):
        return not self.is_in_check(self.turn) and not self.get_all_legal_moves(self.turn)

    def print_board(self):
        # clear_output(wait=True) # ne pas réactiver pour cause de bugs
        print('  a b c d e f g h')
        for row_index, row in enumerate(self.board):
            row_display = f"{8 - row_index} " + ' '.join(
                self.get_piece_representation(piece, row_index, col_index) if piece else self.get_empty_cell(row_index, col_index)
                for col_index, piece in enumerate(row)
            )
            print(row_display)
        print('')

    def get_piece_representation(self, piece, row, col):
        piece_symbol = piece.__class__.__name__[0]
        color = piece.color if piece else None

        # Choisir la couleur de fond
        background_color = "\033[48;5;236m" if (row + col) % 2 == 0 else "\033[48;5;240m"
        # Choisir la couleur du texte
        if color == 'red':
            piece_symbol = piece_symbol.upper()
            text_color = "\033[91m"  # Rouge pour les pièces rouges
        elif color == 'blue':
            piece_symbol = piece_symbol.lower()
            text_color = "\033[94m"  # Bleu pour les pièces bleues
        else:
            piece_symbol = '.'
            text_color = "\033[90m"  # Gris foncé pour les cases vides

        return f"{background_color}{text_color}{piece_symbol}\033[0m"

    def get_empty_cell(self, row, col):
        """Retourne le format de cellule vide avec couleur de fond appropriée."""
        background_color = "\033[48;5;236m" if (row + col) % 2 == 0 else "\033[48;5;240m"
        return f"{background_color} \033[0m"

    def algebraic_to_indices(self, move):
        """Convertir la notation algébrique en indices du plateau."""
        col, row = move[0], move[1]
        col_index = ord(col) - ord('a')
        row_index = 8 - int(row)
        return (row_index, col_index)

    def get_move(self, move):
        """Convertir le mouvement de la notation algébrique en indices du plateau."""
        start, end = move.split()
        return (self.algebraic_to_indices(start), self.algebraic_to_indices(end))

    def get_all_legal_moves(self, color):
        legal_moves = []
        for start_row in range(8):
            for start_col in range(8):
                piece = self.board[start_row][start_col]
                if piece and piece.color == color:
                    for end_row in range(8):
                        for end_col in range(8):
                            if piece.can_move(self.board, (start_row, start_col), (end_row, end_col)):
                                # Vérifier si ce mouvement enlèverait l'échec
                                original_piece = self.board[end_row][end_col]
                                self.board[end_row][end_col] = piece
                                self.board[start_row][start_col] = None
                                if not self.is_in_check(color):
                                    legal_moves.append(((start_row, start_col), (end_row, end_col)))
                                # Revenir en arrière
                                self.board[start_row][start_col] = piece
                                self.board[end_row][end_col] = original_piece
        return legal_moves

    def save_game(self, filename):
        with open(filename, 'w') as file:
            file.write(f"{self.turn}\n")
            for move in self.move_history:
                start, end = move
                start_pos = f"{chr(start[1] + ord('a'))}{8 - start[0]}"
                end_pos = f"{chr(end[1] + ord('a'))}{8 - end[0]}"
                file.write(f"{start_pos} {end_pos}\n")
            for row in self.board:
                file.write(' '.join([f"{p.__class__.__name__[0] if p else '.'}" for p in row]) + '\n')

    def load_game(self, filename):
        with open(filename, 'r') as file:
            lines = file.readlines()
        self.turn = lines[0].strip()
        self.move_history = []
        self.board = [[None for _ in range(8)] for _ in range(8)]
        for r, row in enumerate(lines[2:10]):
            for c, cell in enumerate(row.strip().split()):
                if cell != '.':
                    piece_class = globals().get({'P': Pawn, 'R': Rook, 'C': Cavalier, 'B': Bishop, 'Q': Queen, 'K': King}.get(cell.upper()))
                    self.board[r][c] = piece_class('red') if cell.isupper() else piece_class('blue')
        for move in lines[1:]:
            start, end = self.get_move(move.strip())
            self.move(start, end)

    def print_game_summary(self):
        print("\nRésumé de la partie :")
        print(f"Couleur des pièces : {self.turn} commence la partie.")
        print("Historique des mouvements :")
        for move in self.move_history:
            start, end = move
            start_pos = f"{chr(start[1] + ord('a'))}{8 - start[0]}"
            end_pos = f"{chr(end[1] + ord('a'))}{8 - end[0]}"
            print(f"{start_pos} -> {end_pos}")
        print(f"\nTemps de jeu :")
        print(f"Rouge : {self.player_times['red'].get_time():.2f} secondes")
        print(f"Bleu : {self.player_times['blue'].get_time():.2f} secondes")


class Timer:
    def __init__(self):
        self.start_time = time.time()
        self.elapsed_time = 0

    def start(self):
        self.start_time = time.time()

    def stop(self):
        self.elapsed_time += time.time() - self.start_time

    def reset(self):
        self.elapsed_time = 0
        self.start_time = time.time()

    def get_time(self):
        return self.elapsed_time + (time.time() - self.start_time)

def play_game():
    board = Board()
    current_timer = board.player_times[board.turn]
    current_timer.start()

    while True:
        board.print_board()
        print(f"Tour de {board.turn}")
        
        if board.is_in_check(board.turn):
            print(f"{board.turn} est en échec !")
            if board.is_checkmate():
                print(f"Échec et mat ! {('blue' if board.turn == 'red' else 'red')} gagne !")
                board.print_game_summary()
                break
            else:
                print(f"{board.turn} doit sortir de l'échec.")
        
        move = input("Entrez votre mouvement (par exemple, 'e2 e4') : ")
        try:
            start, end = board.get_move(move)
            if (start, end) in board.get_all_legal_moves(board.turn):
                board.move(start, end)
                current_timer.stop()
                current_timer = board.player_times[board.turn]
                current_timer.start()
            else:
                print("Mouvement illégal.")
        except ValueError as e:
            print(e)
        except Exception as e:
            print(f"Erreur : {e}")

        if board.is_stalemate():
            print("Match nul !")
            board.print_game_summary()
            break
````
